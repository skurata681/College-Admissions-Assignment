```python
## DO NOT MODIFY ##

# Requires: list (representing one student's data)
# Modifies: nothing
# Effects: returns True if all elements in list are float type

def check_row_type(student_data):
    strCount = 0
    for i in range(len(student_data)):
        if isinstance(student_data[i], float):
            strCount += 1
        else:
            return False
    if strCount == len(student_data):
        return True
```

```python
# Requires: list (representing one student's data)
# Modifies: nothing
# Effects: converts list element types to float 

def convert_row_type(student_data):
    
    # convert each element to float
    for i in range(len(student_data)):
        student_data[i] = float(student_data[i])
```

```python
# Requires: list (representing one student's data)
# Modifies: nothing
# Effects: returns student's overall score to 2 decimal places 
# based on weighted average:
    # 40% GPA
    # 30% SAT
    # 20% Curriculum strength
    # 10% Demonstrated interest

def calculate_score(student_data):
    
    # normalize GPA and SAT
    norm_gpa = student_data[0] * 2
    norm_sat = student_data[1] / 160
    curr_strength = student_data[2]
    dem_interest = student_data[3]
    
    # calculate weighted average from student data
    gpa_weight = 0.4
    sat_weight = 0.3
    curr_strength_weight = 0.2
    dem_interest_weight = 0.1
    
    weight_avg = (norm_gpa * gpa_weight) + (norm_sat * sat_weight) + (curr_strength * curr_strength_weight) + (dem_interest * dem_interest_weight)
    weight_avg = round(weight_avg, 2)
    
    return weight_avg
```

```python
# Requires: list (representing one student's data)
# Modifies: nothing
# Effects: returns True if:
    # (1) demonstrated interested score = 0
    # (2) normalized GPA is more than 2 points higher than normalized SAT

def is_outlier(student_data):
    
    dem_interest = student_data[3]
    norm_gpa = student_data[0] * 2
    norm_sat = student_data[1] / 160
    
    # check conditions for outlier
    if ((dem_interest == 0.0) or ((norm_gpa - norm_sat) > 2.0)):
        return True
    
    else:
        return False
    
```

```python
# Requires: list (representing one student's data)
# Modifies: nothing
# Effects: returns True if:
    # (1) student score of 6 or higher
    # (2) student is an outlier

def improved_calculate_score(student_data):
    
    # check improved conditions for chosen student
    if ((calculate_score(student_data) >= 6) or (is_outlier(student_data))):
        return True
        
    else:
        return False
    
```

```python
def main():
    
# Read in data
    dataset_file = open('dataset1.csv', 'r')
    
    # initialize 
    idx_counter = 0
    studentIDs = []
    student_scores_file = open('student_scores.csv', 'w')
    chosen_students_file = open('chosen_students.csv', 'w')
    outliers_file = open('outliers.txt', 'w')
    improved_chosen_file = open('improved_chosen.csv', 'w')
    
    for line in dataset_file:
        # read data into lists
        line = line.replace('\n', '') # line is a string
        list_form = line.split(',') # line is now a list

        # save studentIDs in a list
        studentIDs.append(list_form[0])

        # remove student ID from data
        list_form.pop(0)
        quant_data = list_form # list now only contains quantitative data

# Convert rows to float
        convert_row_type(quant_data)

# Call check_row_type and print error message if False
        if (not check_row_type(quant_data)):
            print('Error: wrong row type')

# Call calculate_score and write scores to a file, 'student_scores.csv' 
# paired with their associated student ID's
        student_scores_file.write(studentIDs[idx_counter] + ',' + str(calculate_score(quant_data)) + '\n')
        

# Write student ID's with scores 6 or higher to a file,'chosen_students.csv'
        if (calculate_score(quant_data) >= 6.0):
            chosen_students_file.write(studentIDs[idx_counter] + ',' + str(calculate_score(quant_data)) + '\n')

# Call is_outlier and write only student ID's to a file,'outliers.txt'
        if is_outlier(quant_data):
            outliers_file.write(studentIDs[idx_counter] + '\n')

# Call improved_calculate_score and write all student information (ID, GPA, SAT, Curr. strength, Dem. interest)
# to new file, 'improved_chosen.csv'
        if improved_calculate_score(quant_data):
            improved_chosen_file.write(studentIDs[idx_counter] + ',' + str(quant_data[0]) + ',' + 
                                      str(quant_data[1]) + ',' + str(quant_data[2]) + ',' + str(quant_data[3]) + '\n')

        idx_counter += 1

    dataset_file.close()
    student_scores_file.close()
    chosen_students_file.close()
    outliers_file.close()
    improved_chosen_file.close()
```

```python
# test
main()
```
